```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

def main() :
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False
    plt.rcParams["figure.figsize"] = (10, 5)
    
    month=["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]
    plt.title("2022년 각 지역별 월평균 기온 추이")
    
    ydata=[]
    nationY=[]
    f=open("Nation.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    for row in data:
        row[2]=float(row[2])
        ydata.append(row[2])
        nationY.append(row[2])
    plt.plot(month, ydata, color="gray", label="전국", linewidth=5)
    f.close()
    
    ydata=[]
    seoulY=[]
    f=open("Seoul.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    for row in data:
        row[2]=float(row[2])
        ydata.append(row[2])
        seoulY.append(row[2])
    plt.plot(month, ydata, color="red", label="서울")
    f.close()
    
    ydata=[]
    daejeonY=[]
    f=open("Daejeon.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    for row in data:
        row[2]=float(row[2])
        ydata.append(row[2])
        daejeonY.append(row[2])
    plt.plot(month, ydata, color="orange", label="대전")
    f.close()
    
    ydata=[]
    busanY=[]
    f=open("Busan.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    for row in data:
        row[2]=float(row[2])
        ydata.append(row[2])
        busanY.append(row[2])
    plt.plot(month, ydata, color="green", label="부산")
    f.close()
    
    ydata=[]
    jejuY=[]
    f=open("Jeju.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    for row in data:
        row[2]=float(row[2])
        ydata.append(row[2])
        jejuY.append(row[2])
    plt.plot(month, ydata, color="blue", label="제주")
    f.close()
    
    plt.xlabel("월")
    plt.ylabel("평균기온 (℃)")
    plt.yticks(range(-5, 35, 5))
    plt.legend()
    plt.show()
    
    for i in range(12) :
        seoulY[i]-=nationY[i]
        daejeonY[i]-=nationY[i]
        busanY[i]-=nationY[i]
        jejuY[i]-=nationY[i]
    plt.plot(month, seoulY, color="red", label="서울")
    plt.plot(month, daejeonY, color="orange", label="대전")
    plt.plot(month, busanY, color="green", label="부산")
    plt.plot(month, jejuY, color="blue", label="제주")
    plt.title("전국 월평균 기온과 각 지역 월평균 기온간의 차이")
    plt.xlabel("월")
    plt.ylabel("각 지역 월평균 기온 - 전국 월평균 기온 ((℃))")
    plt.legend()
    plt.show()
    

if __name__ == "__main__":
    main()
```

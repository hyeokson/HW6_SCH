```python
import csv
import random
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

def dice(list, count) :
    while count :
        num=random.randint(1, 6)
        list.append(num)
        count-=1

def make_csv() :
    dice_100=[]
    dice_1000=[]
    dice_10000=[]
    dice_100000=[]
    
    dice(dice_100, 100)
    dice(dice_1000, 1000)
    dice(dice_10000, 10000)
    dice(dice_100000, 100000)
    
    f=open("random_dice.csv", 'w', encoding="cp949", newline='')
    w=csv.writer(f)
    w.writerow(["dice_100", "dice_1000", "dice_10000", "dice_100000"])
    
    
    idx=0
    while idx<=100000-1 :
        if idx<=100-1 :
            w.writerow([dice_100[idx], dice_1000[idx], dice_10000[idx], dice_100000[idx]])
        elif idx<=1000-1 :
            w.writerow(['', dice_1000[idx], dice_10000[idx], dice_100000[idx]])
        elif idx<=10000-1 :
            w.writerow(['', '', dice_10000[idx], dice_100000[idx]])
        else :
            w.writerow(['', '', '', dice_100000[idx]])
        idx+=1
        
    f.close()

def main() :
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False
    
    dice_100=[]
    dice_1000=[]
    dice_10000=[]
    dice_100000=[]
    
    make_csv()
    
    f=open("random_dice.csv", 'r')
    data=csv.reader(f)
    header=next(data)
    for row in data : 
        if(row[0]!='') :
            dice_100.append(int(row[0]))
        if(row[1]!='') :
            dice_1000.append(int(row[1]))
        if(row[2]!='') :
            dice_10000.append(int(row[2]))
        if(row[3]!='') :
            dice_100000.append(int(row[3]))
    f.close()
    
    x=[1,2,3,4,5,6]
    y=[0, 0, 0, 0, 0, 0]
    plt.subplot(2,2,1)
    plt.title(" 100 ")
    plt.xlabel("  (1~6  )")
    plt.ylabel("")
    plt.xticks([1,2,3,4,5,6])
    for i in range(100) :
        y[dice_100[i]-1]+=1
    for i, v in enumerate(x) :
        plt.text(v, y[i], y[i], color='gray', horizontalalignment='center')
    plt.hist(dice_100, bins=15, color='red')
    
    x=[1,2,3,4,5,6]
    y=[0, 0, 0, 0, 0, 0]
    plt.subplot(2,2,2)
    plt.title(" 1000 ")
    plt.xlabel("  (1~6  )")
    plt.ylabel("")
    plt.xticks([1,2,3,4,5,6])
    for i in range(1000) :
        y[dice_1000[i]-1]+=1
    for i, v in enumerate(x) :
        plt.text(v, y[i], y[i], color='gray', horizontalalignment='center')
    plt.hist(dice_1000, bins=15, color='orange')
    
    x=[1,2,3,4,5,6]
    y=[0, 0, 0, 0, 0, 0]
    plt.subplot(2,2,3)
    plt.title(" 10000 ")
    plt.xlabel("  (1~6  )")
    plt.ylabel("")
    plt.xticks([1,2,3,4,5,6])
    for i in range(10000) :
        y[dice_10000[i]-1]+=1
    for i, v in enumerate(x) :
        plt.text(v, y[i], y[i], color='gray', horizontalalignment='center')
    plt.hist(dice_10000, bins=15, color='green')
    
    x=[1,2,3,4,5,6]
    y=[0, 0, 0, 0, 0, 0]
    plt.subplot(2,2,4)
    plt.title(" 100000 ")
    plt.xlabel("  (1~6  )")
    plt.ylabel("")
    plt.xticks([1,2,3,4,5,6])
    for i in range(100000) :
        y[dice_100000[i]-1]+=1
    for i, v in enumerate(x) :
        plt.text(v, y[i], y[i], color='gray', horizontalalignment='center')
    plt.hist(dice_100000, bins=15, color='blue')
    
    plt.tight_layout()
    plt.show()

        

if __name__=="__main__" :
    main()
```

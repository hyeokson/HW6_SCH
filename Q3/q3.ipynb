```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

def main() :
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False
    
    male=[]
    female=[]
    label=["남자 인구수", "여자 인구수"]
    color=["skyblue", "hotpink"]
    
    f=open("PopulationByGender.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    for row in data :
        for i in range(1, 41, 12) :
            row[i+2]=row[i+2].replace(',', '')
            row[i+3]=row[i+3].replace(',', '')
            male.append(int(row[i+2]))
            female.append(int(row[i+3]))
    f.close()
    
    for i in range(4) :
        plt.subplot(2,2,i+1)
        plt.title(str(2013+3*i)+"년 제주도 남녀 인구 비율")
        plt.pie([male[i], female[i]], labels=label, autopct="%.3f%%", colors=color)
    
    plt.tight_layout()
    plt.show()
    
    
    
if __name__=="__main__" :
    main()
```

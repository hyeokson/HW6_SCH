```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

def main() :
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False
    
    getOn={}
    getOff={}
    getOnOff={}
    
    f=open("Subway.csv", 'r', encoding="ANSI")
    data=csv.reader(f)
    header=next(data)
    next(data)
    for row in data :
        if row[3] not in getOn :
            getOn[row[3]]=int(row[10])+int(row[12])
            getOff[row[3]]=int(row[11])+int(row[13])
            getOnOff[row[3]]=int(row[10])+int(row[11])+int(row[12])+int(row[13])
        else :
            getOn[row[3]]+=int(row[10])+int(row[12])
            getOff[row[3]]+=int(row[11])+int(row[13])
            getOnOff[row[3]]+=int(row[10])+int(row[11])+int(row[12])+int(row[13])
    f.close()
    
    sorted_getOn=sorted(getOn.items(), key=lambda item: item[1], reverse=True)
    xdata=[]
    ydata=[]
    for i in range(30) :
        xdata.append(sorted_getOn[i][0])
        ydata.append(sorted_getOn[i][1])
    dict(sorted_getOn)
    plt.subplot(3, 1, 1)
    plt.xticks(rotation=90)
    plt.rcParams["figure.figsize"] = (20, 10)
    plt.title("출근시간 최대 승차역 30개")
    plt.xlabel("승차역")
    plt.ylabel("승차한 승객수")
    plt.bar(xdata, ydata)
    
    sorted_getOff=sorted(getOff.items(), key=lambda item: item[1], reverse=True)
    xdata=[]
    ydata=[]
    for i in range(30) :
        xdata.append(sorted_getOff[i][0])
        ydata.append(sorted_getOff[i][1])
    sorted_getOff=sorted(getOff.items(), key=lambda item: item[1], reverse=True)
    plt.subplot(3,1,2)
    plt.xticks(rotation=90)
    plt.rcParams["figure.figsize"] = (20, 10)
    plt.title("출근시간 최대 하차역 30개")
    plt.xlabel("하차역")
    plt.ylabel("하차한 승객수")
    plt.bar(xdata, ydata)
    
    sorted_getOnOff=sorted(getOnOff.items(), key=lambda item: item[1], reverse=True)
    xdata=[]
    ydata=[]
    for i in range(30) :
        xdata.append(sorted_getOnOff[i][0])
        ydata.append(sorted_getOnOff[i][1])
    sorted_getOnOff=sorted(getOnOff.items(), key=lambda item: item[1], reverse=True)
    plt.subplot(3,1,3)
    plt.xticks(rotation=90)
    plt.rcParams["figure.figsize"] = (20, 10)
    plt.title("출근시간 최대 승하차역 30개")
    plt.xlabel("승하차역")
    plt.ylabel("승하차한 승객수")
    plt.bar(xdata, ydata)
    
    plt.tight_layout()
    plt.show()

if __name__=="__main__" :
    main()
```
